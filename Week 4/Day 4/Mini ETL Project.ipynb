```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("MiniETLProject").getOrCreate()
```

```python
raw_data = [
    (1, "Laptop", "Electronics", 50000, 1, "2025-01-10", "Hyderabad"),
    (2, "Mobile", "Electronics", 30000, 2, "2025-01-11", "Bangalore"),
    (3, None, "Electronics", 45000, 1, "2025-01-12", "Chennai"),
    (4, "Shoes", "Fashion", 3000, 2, "2025-01-13", "Hyderabad"),
    (5, "Watch", "Fashion", None, 1, "2025-01-14", "Mumbai"),
    (6, "Refrigerator", "Appliances", 35000, 1, "2025-01-15", None),
    (7, "AC", "Appliances", 40000, 1, "2025-01-16", "Bangalore"),
    (8, "Mobile", "Electronics", 25000, 3, "2025-01-17", "Pune"),
    (9, "Shoes", "Fashion", 3500, -2, "2025-01-18", "Chennai"),
    (10, "Laptop", "Electronics", 60000, 1, "2025-01-19", "Kolkata"),
    (11, "Laptop", "Electronics", 60000, 1, "2025-01-19", "Kolkata")
]

columns = [
    "order_id",
    "product",
    "category",
    "price",
    "quantity",
    "order_date",
    "city"
]

raw_df = spark.createDataFrame(raw_data, columns)

raw_df.display()
```

```python
raw_df.printSchema()
```

```python
no_dupes_df = raw_df.dropDuplicates()

no_dupes_df.display()
```

```python
clean_df = no_dupes_df.dropna(
    subset=["product", "price"]
)

clean_df.display()
```

```python
clean_df = clean_df.fillna(
    {"city": "Unknown"}
)

clean_df.display()
```

```python
clean_df = clean_df.filter(
    col("quantity") > 0
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "order_date",
    to_date(col("order_date"))
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "total_amount",
    col("price") * col("quantity")
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "value_segment",
    when(col("total_amount") > 50000, "High")
    .when(col("total_amount").between(20000, 50000), "Medium")
    .otherwise("Low")
)

clean_df.display()
```

```python
city_summary_df = clean_df.groupBy("city") \
    .agg(
        sum("total_amount").alias("total_revenue"),
        count("order_id").alias("total_orders"),
        round(avg("total_amount"), 2).alias("avg_order_value")
    ) \
    .orderBy(col("total_revenue").desc())

city_summary_df.display()
```

```python
category_summary_df = clean_df.groupBy("category") \
    .agg(
        sum("total_amount").alias("total_revenue"),
        count("order_id").alias("total_orders")
    ) \
    .orderBy(col("total_revenue").desc())

category_summary_df.display()
```

```python
clean_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("clean_sales_data")
```

```python
city_summary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("city_sales_summary")
```

```python
category_summary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("category_sales_summary")
```
