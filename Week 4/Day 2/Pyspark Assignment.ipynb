```python
from pyspark.sql.functions import *

employee_data = [
    (101,"Sravan","Data Engineer","IT",75000,"Hyderabad",28,"2021-05-10","Male"),
    (102,"Ravi","Software Engineer","IT",68000,"Bangalore",30,"2020-03-15","Male"),
    (103,"Priya","Data Analyst","Analytics",62000,"Chennai",26,"2022-01-12","Female"),
    (104,"Kiran","Manager","HR",90000,"Mumbai",35,"2018-07-19","Male"),
    (105,"Anjali","HR Executive","HR",45000,"Pune",24,"2023-02-20","Female"),
    (106,"Vikram","Data Scientist","Analytics",98000,"Delhi",32,"2019-11-25","Male"),
    (107,"Sneha","Developer","IT",71000,"Hyderabad",27,"2021-08-17","Female"),
    (108,"Rahul","Tester","QA",55000,"Chennai",29,"2020-06-10","Male"),
    (109,"Meena","QA Lead","QA",83000,"Bangalore",33,"2017-09-14","Female"),
    (110,"Arjun","Support Engineer","Support",50000,"Pune",31,"2022-04-11","Male")
]

columns = ["emp_id","name","designation","department","salary","city","age","joining_date","gender"]

emp_df = spark.createDataFrame(employee_data, columns)

emp_df.display()
```

##11. Joins

```python
department_data = [
    ("IT","John"),
    ("HR","Smith"),
    ("QA","David"),
    ("Analytics","Kevin"),
    ("Support","Robert")
]

dept_columns = ["department","manager"]

dept_df = spark.createDataFrame(department_data, dept_columns)

dept_df.display()
```

```python
emp_df.join(dept_df, "department", "inner").display()

emp_df.join(dept_df, "department", "left").display()

emp_df.join(dept_df, "department", "right").display()

emp_df.join(dept_df, "department", "outer").display()

emp_df.join(dept_df, "department") \
      .select("emp_id", "name", "department", "manager") \
      .display()
```

##12. UNION

```python
new_emp_data = [
    (111,"Teja","Developer","IT",72000,"Hyderabad",26,"2023-01-10","Male"),
    (112,"Divya","Analyst","Analytics",65000,"Bangalore",25,"2022-11-05","Female")
]

new_emp_df = spark.createDataFrame(new_emp_data, columns)

union_df = emp_df.union(new_emp_df)

union_df.display()

print(union_df.count())

union_df.dropDuplicates().display()

union_df.filter(col("department") == "IT").display()

union_df.orderBy(col("salary").desc()).display()
```

##13. UNIONBYNAME

```python
shuffled_df = new_emp_df.select(
    "name",
    "emp_id",
    "department",
    "designation",
    "city",
    "salary",
    "age",
    "joining_date",
    "gender"
)

emp_df.unionByName(shuffled_df).display()

extra_df = new_emp_df.withColumn("bonus", lit(5000))

emp_df.unionByName(
    extra_df,
    allowMissingColumns=True
).display()

emp_df.unionByName(new_emp_df).display()

emp_df.unionByName(
    extra_df,
    allowMissingColumns=True
).display()

emp_df.unionByName(
    extra_df,
    allowMissingColumns=True
).display()
```

##16. EXPLODE

```python
skills_data = [
    (101, ["Python", "Spark", "Azure"]),
    (102, ["Java", "SQL"]),
    (103, ["Power BI", "SQL"])
]

skills_columns = ["emp_id", "skills"]

skills_df = spark.createDataFrame(skills_data, skills_columns)

skills_df.select("emp_id", explode("skills").alias("skill")).display()

skills_df.select(explode("skills").alias("skill")).count()

skills_df.select("emp_id", explode("skills").alias("skill")) \
         .filter(col("skill") == "Python") \
         .display()

skills_df.select(explode("skills").alias("skill")) \
         .distinct() \
         .display()

skills_df.select(explode("skills").alias("skill")) \
         .groupBy("skill") \
         .count() \
         .display()
```

##18. CONCAT / CONCAT_WS

```python
emp_df.withColumn(
    "name_department",
    concat(col("name"), col("department"))
).display()

emp_df.withColumn(
    "employee_details",
    concat_ws(" | ",
              col("name"),
              col("designation"),
              col("department"))
).display()

emp_df.withColumn(
    "city_department",
    concat_ws("-", col("city"), col("department"))
).display()

emp_df.withColumn(
    "employee_label",
    concat_ws("_", col("name"), col("emp_id"))
).display()

emp_df.withColumn(
    "name_designation",
    concat_ws(" - ", col("name"), col("designation"))
).display()
```

##22. WHEN / OTHERWISE

```python
emp_df.withColumn(
    "salary_category",
    when(col("salary") > 70000, "High")
    .otherwise("Low")
).display()

emp_df.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .otherwise("Senior")
).display()

emp_df.withColumn(
    "experience_level",
    when(col("age") > 30, "Experienced")
    .otherwise("Junior")
).display()

emp_df.withColumn(
    "bonus_eligible",
    when(col("salary") > 70000, "Yes")
    .otherwise("No")
).display()

emp_df.withColumn(
    "dept_category",
    when(col("department") == "IT", "Technical")
    .otherwise("Non Technical")
).display()
```

##23. SUBSTRING

```python
emp_df.withColumn(
    "name_prefix",
    substring("name", 1, 3)
).display()

emp_df.withColumn(
    "joining_year",
    substring("joining_date", 1, 4)
).display()

emp_df.withColumn(
    "dept_code",
    substring("department", 1, 2)
).display()

emp_df.withColumn(
    "designation_end",
    substring(col("designation"),
              length(col("designation")) - 2,
              3)
).display()

emp_df.withColumn(
    "employee_code",
    concat(
        substring("name",1,3),
        col("emp_id")
    )
).display()
```

##24. REGEXP_REPLACE

```python
emp_df.withColumn(
    "designation",
    regexp_replace("designation", " ", "_")
).display()

emp_df.withColumn(
    "name",
    regexp_replace("name", "[AEIOUaeiou]", "")
).display()

emp_df.withColumn(
    "city",
    regexp_replace("city", "Hyderabad", "HYD")
).display()

emp_df.withColumn(
    "designation",
    regexp_replace("designation", "[^a-zA-Z ]", "")
).display()

emp_df.withColumn(
    "department",
    regexp_replace("department", "IT", "Information Technology")
).display()
```

##28. PIVOT

##30. WINDOW FUNCTIONS

```python
from pyspark.sql.window import Window

salary_window = Window.orderBy(col("salary").desc())

dept_window = Window.partitionBy("department") \
                    .orderBy(col("salary").desc())

emp_df.withColumn(
    "rank",
    rank().over(salary_window)
).display()

emp_df.withColumn(
    "dense_rank",
    dense_rank().over(dept_window)
).display()

emp_df.withColumn(
    "row_number",
    row_number().over(salary_window)
).display()

emp_df.withColumn(
    "lead_salary",
    lead("salary").over(salary_window)
).display()

emp_df.withColumn(
    "lag_salary",
    lag("salary").over(salary_window)
).display()
```
