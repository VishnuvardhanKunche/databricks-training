```python
from pyspark.sql.functions import *

employee_data = [
    (101,"Sravan","Data Engineer","IT",75000,"Hyderabad",28,"2021-05-10","Male"),
    (102,"Ravi","Software Engineer","IT",68000,"Bangalore",30,"2020-03-15","Male"),
    (103,"Priya","Data Analyst","Analytics",62000,"Chennai",26,"2022-01-12","Female"),
    (104,"Kiran","Manager","HR",90000,"Mumbai",35,"2018-07-19","Male"),
    (105,"Anjali","HR Executive","HR",45000,"Pune",24,"2023-02-20","Female"),
    (106,"Vikram","Data Scientist","Analytics",98000,"Delhi",32,"2019-11-25","Male"),
    (107,"Sneha","Developer","IT",71000,"Hyderabad",27,"2021-08-17","Female"),
    (108,"Rahul","Tester","QA",55000,"Chennai",29,"2020-06-10","Male"),
    (109,"Meena","QA Lead","QA",83000,"Bangalore",33,"2017-09-14","Female"),
    (110,"Arjun","Support Engineer","Support",50000,"Pune",31,"2022-04-11","Male")
]

columns = ["emp_id","name","designation","department","salary","city","age","joining_date","gender"]

emp_df = spark.createDataFrame(employee_data, columns)

emp_df.display()
```

##REPARTITION

```python
emp_df.repartition(4).display()

emp_df.repartition("department").display()

large_df = emp_df.repartition(8)
```

##COALESCE

```python
emp_df.coalesce(2).display()

emp_df.repartition(2).display()

emp_df.coalesce(1).display()
```

```python
department_data = [
    ("IT","John"),
    ("HR","Smith"),
    ("QA","David"),
    ("Analytics","Kevin"),
    ("Support","Robert")
]

dept_columns = ["department","manager"]

dept_df = spark.createDataFrame(
    department_data,
    dept_columns
)

dept_df.display()
```

##CROSSJOIN

```python
emp_df.crossJoin(dept_df).display()

emp_df.crossJoin(dept_df).display()

print(emp_df.crossJoin(dept_df).count())

emp_df.select("city").distinct() \
      .crossJoin(
          dept_df.select("department")
      ) \
      .display()

emp_df.crossJoin(dept_df).display()
```

##EXPLODE_OUTER

```python
skills_data = [
    (101, ["Python", "Spark", "Azure"]),
    (102, ["Java", "SQL"]),
    (103, ["Power BI", "SQL"])
]

skills_columns = ["emp_id", "skills"]

skills_df = spark.createDataFrame(
    skills_data,
    skills_columns
)

skills_df.display()
```

```python
skills_null_data = [
    (101, ["Python", "Spark"]),
    (102, None),
    (103, [])
]

skills_null_columns = ["emp_id", "skills"]

skills_null_df = spark.createDataFrame(
    skills_null_data,
    skills_null_columns
)

skills_null_df.display()
```

```python
skills_null_df = spark.createDataFrame([
    (101, ["Python", "Spark"]),
    (102, None),
    (103, [])
], ["emp_id", "skills"])

skills_null_df.select(
    "emp_id",
    explode_outer("skills").alias("skill")
).display()

skills_null_df.select(
    "emp_id",
    explode("skills").alias("skill")
).display()

skills_null_df.select(
    "emp_id",
    explode_outer("skills").alias("skill")
).display()

skills_null_df.select(
    "emp_id",
    explode_outer("skills").alias("skill")
).display()

skills_null_df.select(
    "emp_id",
    explode_outer("skills").alias("skill")
).display()
```
