```python
from pyspark.sql.functions import *

student_data = [
    (101,"Rahul","CSE",85,"Hyderabad"),
    (102,"Priya","ECE",92,"Bangalore"),
    (103,None,"CSE",78,"Chennai"),
    (104,"Anjali","IT",None,"Mumbai"),
    (105,"Kiran","cse",105,"Hyderabad"),
    (106,"Vikram","ECE",-5,"Pune"),
    (107,"Sneha","IT",88,None),
    (108,"Arjun","CSE",65,"Hyd"),
    (109,"Meena","ECE",72,"Bangalore"),
    (110,"Rahul","CSE",85,"Hyderabad")
]

columns = [
    "student_id",
    "student_name",
    "department",
    "marks",
    "city"
]

students_df = spark.createDataFrame(
    student_data,
    columns
)

students_df.display()
```

```python
students_df.printSchema()
```

```python
clean_df = students_df.dropDuplicates()

clean_df.display()
```

```python
clean_df = clean_df.dropna(
    subset=["student_name"]
)

clean_df.display()
```

```python
clean_df = clean_df.fillna(
    {"marks":0}
)

clean_df.display()
```

```python
clean_df = clean_df.fillna(
    {"city":"Unknown"}
)

clean_df.display()
```

```python
clean_df = clean_df.filter(
    (col("marks") >= 0) &
    (col("marks") <= 100)
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "city",
    when(col("city") == "Hyd", "Hyderabad")
    .otherwise(col("city"))
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "grade",
    when(col("marks") >= 90,"A")
    .when(col("marks") >= 75,"B")
    .when(col("marks") >= 60,"C")
    .otherwise("D")
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "result",
    when(col("marks") >= 40,"Pass")
    .otherwise("Fail")
)

clean_df.display()
```

```python
clean_df = clean_df.withColumn(
    "performance",
    when(col("marks") >= 85,"Excellent")
    .when(col("marks") >= 70,"Good")
    .otherwise("Average")
)

clean_df.display()
```

```python
clean_df = clean_df.withColumnRenamed(
    "marks",
    "student_marks"
).withColumnRenamed(
    "city",
    "student_city"
)

clean_df.display()
```

```python
dept_report = clean_df.groupBy(
    "department"
).agg(
    round(avg("student_marks"),2).alias("avg_marks"),
    count("*").alias("students")
)

dept_report.display()
```

```python
city_report = clean_df.groupBy(
    "student_city"
).count()

city_report.display()
```

```python
clean_df.groupBy(
    "grade"
).count().display()
```

```python
clean_df.groupBy(
    "performance"
).count().display()
```

```python
clean_df.groupBy(
    "department"
).pivot(
    "grade"
).count().display()
```

```python
clean_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("clean_student_data")
```

```python
print("Raw Records :", students_df.count())

print("Clean Records :", clean_df.count())

print(
    "Removed Records :",
    students_df.count() - clean_df.count()
)
```
