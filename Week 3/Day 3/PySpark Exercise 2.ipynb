## WITHCOLUMN

```python


df.withColumn("bonus", col("salary") * 0.10).display()

df.withColumn("annual_salary", col("salary") * 12).display()

df.withColumn("tax", col("salary") * 0.05).display()

df.withColumn("updated_salary", col("salary") + 5000).display()

df.withColumn(
    "salary_category",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 50000, "Medium")
    .otherwise("Low")
).display()

df.withColumn(
    "age_group",
    when(col("age") < 25, "Young")
    .otherwise("Adult")
).display()

df.withColumn(
    "location",
    concat_ws("-", col("city"), col("department"))
).display()

df.withColumn(
    "increment_salary",
    col("salary") * 1.15
).display()

df.withColumn(
    "experience_status",
    when(year(to_date(col("joining_date"))) < 2022, "Experienced")
    .otherwise("New")
).display()

df.withColumn(
    "name_length",
    length(col("emp_name"))
).display()

df.withColumn(
    "is_high_salary",
    when(col("salary") > 80000, True)
    .otherwise(False)
).display()

df.withColumn(
    "joining_year",
    year(to_date(col("joining_date")))
).display()

df.withColumn(
    "salary_after_tax",
    col("salary") - (col("salary") * 0.05)
).display()

df.withColumn(
    "department_code",
    substring(col("department"), 1, 3)
).display()

df.withColumn(
    "double_salary",
    col("salary") * 2
).display()
```
