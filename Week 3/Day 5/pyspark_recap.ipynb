```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("PySparkSyntax").getOrCreate()

data = [
    (101,"Rahul","IT",65000,25,"Hyderabad"),
    (102,"Priya","HR",55000,28,"Bangalore"),
    (103,"Arjun","IT",70000,30,"Chennai"),
    (104,"Sneha","Finance",80000,27,"Mumbai"),
    (105,"Kiran","HR",50000,24,"Hyderabad")
]

columns = [
    "emp_id",
    "emp_name",
    "department",
    "salary",
    "age",
    "city"
]

df = spark.createDataFrame(data, columns)

df.display()
```

```python
df.select("emp_name","salary").display()
```

```python
df.select(
    col("emp_name").alias("Employee"),
    col("salary").alias("Salary")
).display()
```

```python
df.filter(col("salary")>60000).display()
```

```python
df.where(col("department")=="IT").display()
```

```python
df.withColumn(
    "bonus",
    col("salary")*0.10
).display()
```

```python
df.withColumnRenamed("emp_name","employee name").display()
```

```python
df.drop(col("age")).display()
```

```python
df.select("department").distinct().display()
```

```python
df.dropDuplicates().display()
```

```python
df.orderBy("salary").display()
```

```python
df.orderBy(col("salary").desc()).display()
```

```python
df.limit(3).display()
```

```python
df.groupBy("department").sum("salary").display()
```

```python
df.groupBy("department").agg(
    avg("salary").alias("avg_salary"),
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary")
).display()
```

```python
dept = [
    ("IT","John"),
    ("HR","Smith"),
    ("Finance","Kevin")
]

dept_df = spark.createDataFrame(
    dept,
    ["department","manager"]
)

df.join(
    dept_df,
    "department",
    "inner"
).display()
```

```python
df.union(df).display()
```

```python
df.unionByName(df).display()
```

```python
df.fillna({
    "city":"Unknown"
}).display()
```

```python
df.dropna().display()
```

```python
df.replace(
    "Hyderabad",
    "Hyd",
    subset=["city"]
).display()
```

```python
df.withColumn(
    "salary",
    col("salary").cast("double")
).display()
```

```python
df.withColumn(
    "grade",
    when(col("salary") >= 70000,"High")
    .otherwise("Normal")
).display()
```

```python
df.withColumn(
    "emp_name",
    upper("emp_name")
).display()
```

```python
df.withColumn(
    "today",
    current_date()
).display()
```

```python
df.withColumn(
    "city_array",
    split(col("city"),"a")
).display()
```

```python
skills = [
    (101,["Python","Spark"]),
    (102,["Java","SQL"])
]

skills_df = spark.createDataFrame(
    skills,
    ["emp_id","skills"]
)

skills_df.select(
    "emp_id",
    explode("skills")
).display()
```

```python
skills_df.filter(
    array_contains("skills","Python")
).display()
```

```python
df.groupBy("city") \
  .pivot("department") \
  .sum("salary") \
  .display()
```

```python
df.cube(
    "city",
    "department"
).sum("salary").display()
```

```python
df.rollup(
    "city",
    "department"
).sum("salary").display()
```

```python
df.crossJoin(dept_df).display()
```

```python
from pyspark.sql.functions import broadcast

df.join(
    broadcast(dept_df),
    "department"
).display()
```

```python
df.exceptAll(df.limit(2)).display()
```

```python
df.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("employee_data")
```
